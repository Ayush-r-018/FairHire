```python
# ==========================================
# Notebook 02 : Data Preprocessing
# ==========================================

import pandas as pd
import numpy as np

import os
import sys

# Add project root to Python path
sys.path.append(os.path.abspath(".."))

from src.utils import project_name

print(f"Project: {project_name()}")
print("Notebook 02 - Data Preprocessing")
```

```python
# Load datasets

job_df = pd.read_csv("../data/raw/job_description_dataset.csv")
resume_df = pd.read_csv("../data/raw/resume_dataset.csv")

print("Job Dataset Shape :", job_df.shape)
print("Resume Dataset Shape :", resume_df.shape)
```

```python
display(job_df.head())

display(resume_df.head())
```

```python
# 4. Missing Values
print("Missing Values in Job Dataset")
display(job_df.isnull().sum())

print("\nMissing Values in Resume Dataset")
display(resume_df.isnull().sum())
```

```python
# 5. Remove Missing Values
job_df = job_df.dropna()

resume_df = resume_df.dropna()

print("Job Dataset Shape :", job_df.shape)
print("Resume Dataset Shape :", resume_df.shape)
```

```python
# 6. Duplicate Analysis
print("Duplicate Job Records :", job_df.duplicated().sum())

print("Duplicate Resume Records :", resume_df.duplicated().sum())
```

```python
# 7. Download NLTK Resources
import nltk

nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
# 8 . Import Preprocessing Module
from src.preprocessing import clean_text

print("Preprocessing module imported successfully.")
```

```python
# 9. Apply Preprocessing
resume_df["Clean_Resume"] = resume_df["Resume"].apply(clean_text)

resume_df["Clean_Job_Description"] = (
    resume_df["Job_Description"].apply(clean_text)
)
```

```python
# 9.5 Clean Job Dataset

job_df["Clean_Skills"] = job_df["Skills"].apply(clean_text)

job_df["Clean_Responsibilities"] = job_df["Responsibilities"].apply(clean_text)

job_df["Clean_Keywords"] = job_df["Keywords"].apply(clean_text)
```

```python
job_df["Clean_Job_Text"] = (
    job_df["Title"].fillna("") + " " +
    job_df["Clean_Skills"] + " " +
    job_df["Clean_Responsibilities"] + " " +
    job_df["Clean_Keywords"]
)
```

```python
# 10. Display Original and Cleaned Resume
print("=" * 80)
print("ORIGINAL RESUME")
print("=" * 80)

print(resume_df.loc[0, "Resume"][:800])

print("\n")

print("=" * 80)
print("CLEANED RESUME")
print("=" * 80)

print(resume_df.loc[0, "Clean_Resume"][:800])
```

```python
# Create processed directory
os.makedirs("../data/processed", exist_ok=True)

# Save cleaned datasets
resume_df.to_csv(
    "../data/processed/cleaned_resume_dataset.csv",
    index=False
)

job_df.to_csv(
    "../data/processed/cleaned_job_dataset.csv",
    index=False
)

print("✅ Cleaned resume dataset saved.")
print("✅ Cleaned job dataset saved.")
```
